```python
import tensorflow
import pandas as pd
import string
```

```python
#saving chat text file to string
with open(r"C:\Users\Enrique\\Desktop\chat_file.txt", 'r', encoding='utf8') as file:
    corpus = file.readlines()
```

```python
#Removing stop words
def remove_stop_words(corpus):
    stop_words = ['is', 'a', 'will', 'be']
    results = []
    for text in corpus:
        tmp = text.split(' ')
        for stop_word in stop_words:
            if stop_word in tmp:
                tmp.remove(stop_word)
        results.append(" ".join(tmp))
    
    return results
```

```python
#Calling Remove Stop Word Function
corpus = remove_stop_words(corpus)
```

```python
#creating word vector with unique words (no stop words)
words = []
for text in corpus:
    for word in text.split(' '):
        words.append(word[21:]) #removes timestamp but not the username

words = set(words)
```

```python
words
```

```python
word2int ={}


for i,word in enumerate(words):
    word2int[word] = i

sentences = []
for sentence in corpus:
    sentences.append(sentence.split())
    
WINDOW_SIZE = 2

data = []
for sentence in sentences:
    for idx, word in enumerate(sentence):
        for neighbor in sentence[max(idx - WINDOW_SIZE, 0) : min(idx + WINDOW_SIZE, len(sentence)) + 1] : 
            if neighbor != word:
                data.append([word, neighbor])
```

```python
for text in corpus:
    print(text)

df = pd.DataFrame(data, columns = ['input', 'label'])
```
